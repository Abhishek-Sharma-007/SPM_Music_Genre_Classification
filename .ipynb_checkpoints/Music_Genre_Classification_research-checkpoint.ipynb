```python
DATASET_PATH = "/content/Data/genres_original"
JSON_PATH = "data_10.json"
SAMPLE_RATE = 22050
TRACK_DURATION = 30  # measured in seconds
SAMPLES_PER_TRACK = SAMPLE_RATE * TRACK_DURATION
```

```python
import os
import json
import math
import random

import numpy as np
import librosa
import librosa.display
import tensorflow.keras as keras
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
def save_mfcc(dataset_path, json_path, num_mfcc=13, n_fft=2048, hop_length=512, num_segments=5):
    """Extracts MFCCs from music dataset and saves them into a json file along with genre labels."""

    # Dictionary to store mapping, labels, and MFCCs
    data = {
        "mapping": [],
        "labels": [],
        "mfcc": []
    }

    samples_per_segment = int(SAMPLES_PER_TRACK / num_segments)
    num_mfcc_vectors_per_segment = math.ceil(samples_per_segment / hop_length)

    # Loop through all genre sub-folders
    for i, (dirpath, dirnames, filenames) in enumerate(os.walk(dataset_path)):

        # Ensure we're processing a genre sub-folder level
        if dirpath != dataset_path:

            # Save genre label (i.e., sub-folder name) in the mapping
            semantic_label = dirpath.split("/")[-1]
            data["mapping"].append(semantic_label)
            print("\nProcessing: {}".format(semantic_label))

            # Process all audio files in genre sub-directory
            for f in filenames:

                # Ensure only .wav files are processed
                if f.endswith('.wav'):

                    # Load audio file
                    file_path = os.path.join(dirpath, f)
                    signal, sample_rate = librosa.load(file_path, sr=SAMPLE_RATE)

                    # Process all segments of audio file
                    for d in range(num_segments):

                        # Calculate start and finish sample for current segment
                        start = samples_per_segment * d
                        finish = start + samples_per_segment

                        # Extract mfcc
                        mfcc = librosa.feature.mfcc(y=signal[start:finish], sr=SAMPLE_RATE, n_mfcc=num_mfcc, n_fft=n_fft, hop_length=hop_length)
                        mfcc = mfcc.T

                        # Store only MFCC feature with the expected number of vectors
                        if len(mfcc) == num_mfcc_vectors_per_segment:
                            data["mfcc"].append(mfcc.tolist())
                            data["labels"].append(i-1)
                            print("{}, segment:{}".format(file_path, d+1))

    # Save MFCCs to json file
    with open(json_path, "w") as fp:
        json.dump(data, fp, indent=4)
```

```python
save_mfcc(DATASET_PATH, JSON_PATH, num_segments=10)
```

```python
import json
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow import keras

def load_data(data_path):
    with open(data_path, "r") as f:
        data = json.load(f)

    # convert lists to numpy arrays
    X = np.array(data["mfcc"])
    y = np.array(data["labels"])

    print("Data succesfully loaded!")
    return  X, y
```

```python
# Ensure you define DATA_PATH before calling the function or replace DATA_PATH with the actual path string.
X, y = load_data(DATA_PATH)
```

```python
def display_data(X, y, num_samples=5):
    print(f"Shape of X: {X.shape}")
    print(f"Shape of y: {y.shape}\n")

    print(f"First {num_samples} entries of X:")
    print(X[:num_samples])

    print(f"\nFirst {num_samples} entries of y:")
    print(y[:num_samples])

# Call the display function
display_data(X, y)
```

```python
# Check the unique labels to determine the number of classes
num_classes = len(np.unique(y))
print(num_classes)
```

```python
# create train/test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
def transform_labels(y):
    # Create a dictionary to map original labels to new labels
    unique_labels = np.unique(y)
    label_mapping = {original_label: idx for idx, original_label in enumerate(unique_labels)}

    # Use the map to transform y
    return np.array([label_mapping[label] for label in y])

y = transform_labels(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

### Building 1st Sequential model

```python
# build network topology
model_1 = keras.Sequential([
    # input layer
    keras.layers.Flatten(input_shape=(X.shape[1], X.shape[2])),
    # 1st dense layer
    keras.layers.Dense(512, activation='relu'),
    # 2nd dense layer
    keras.layers.Dense(256, activation='relu'),
    # 3rd dense layer
    keras.layers.Dense(64, activation='relu'),
    # output layer
    keras.layers.Dense(num_classes, activation='softmax')  # Adjusted this line
])
```

```python
# compile model
optimiser = keras.optimizers.Adam(learning_rate=0.0001)
model_1.compile(optimizer=optimiser,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# train model
history = model_1.fit(X_train, y_train, validation_data=(X_test, y_test), batch_size=32, epochs=50)
```

```python
def plot_history(history):

    fig, axs = plt.subplots(2)

    # create accuracy sublpot
    axs[0].plot(history.history["accuracy"], label="train accuracy")
    axs[0].plot(history.history["val_accuracy"], label="test accuracy")
    axs[0].set_ylabel("Accuracy")
    axs[0].legend(loc="lower right")
    axs[0].set_title("Accuracy eval")

    # create error sublpot
    axs[1].plot(history.history["loss"], label="train error")
    axs[1].plot(history.history["val_loss"], label="test error")
    axs[1].set_ylabel("Error")
    axs[1].set_xlabel("Epoch")
    axs[1].legend(loc="upper right")
    axs[1].set_title("Error eval")

    plt.show()
```

Adjusting the learning rate can have a significant impact on training: too high can cause divergence, too low can cause slow convergence.

5. **Loss Function (from your previous code):**
    - **Sparse Categorical Crossentropy**: This is appropriate when the target values are integers (as opposed to one-hot encoded vectors). It measures the dissimilarity between the predicted probability distribution and the true distribution.

6. **Metrics (from your previous code):**
    - **Accuracy**: This metric calculates how often predictions equal labels. It's a common metric for classification problems.

# Building 3rd Sequential Model

```python
import keras
from keras.layers import Flatten, Dense, Dropout, BatchNormalization, LeakyReLU
from keras.optimizers import RMSprop , Adam , SGD , Adagrad
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
```

```python
# Define the model using Keras Sequential API
model_3 = keras.Sequential([

    # input layer
    Flatten(input_shape=(X.shape[1], X.shape[2])),

    # 1st dense layer
    Dense(512, kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001)),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(0.4),

    # 2nd dense layer
    Dense(256, kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001)),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(0.4),

    # 3rd dense layer
    Dense(64, kernel_initializer='he_normal', kernel_regularizer=keras.regularizers.l2(0.001)),
    LeakyReLU(),
    BatchNormalization(),
    Dropout(0.4),

    # output layer
    Dense(10, activation='softmax')
])
```

```python
# Use the RMSprop optimizer
optimiser = RMSprop(learning_rate=0.0001)

# Compile the model
model_3.compile(optimizer=optimiser,
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

# Define learning rate scheduler and early stopping
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
```

```python
# Train the model with validation data and callbacks
history = model_3.fit(X_train, y_train,
                      validation_data=(X_test, y_test),
                      batch_size=32, epochs=100,
                      callbacks=[lr_scheduler, early_stopping])
```

```python
# plot accuracy and error as a function of the epochs
plot_history(history)
```

```python
# Extract best training and validation accuracies and loss
best_train_acc = max(history.history['accuracy'])
best_train_loss = min(history.history['loss'])
best_val_acc = max(history.history['val_accuracy'])
best_val_loss = min(history.history['val_loss'])

print("\nBest Training Accuracy: {:.4f}".format(best_train_acc))
print("Best Training Loss: {:.4f}".format(best_train_loss))
print("Best Validation Accuracy: {:.4f}".format(best_val_acc))
print("Best Validation Loss: {:.4f}".format(best_val_loss))
```

```python
# TensorFlow and Keras imports
import tensorflow as tf
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
from tensorflow.keras.regularizers import l2
```

```python
model_9 = keras.Sequential([

    # Assuming MFCCs, start with convolutional layers
    keras.layers.Reshape((X.shape[1], X.shape[2], 1), input_shape=(X.shape[1], X.shape[2])),
    keras.layers.Conv2D(32, (3, 3), activation='relu'),
    keras.layers.MaxPooling2D((2, 2)),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.3),

    # Flattening the output to feed into dense layers
    keras.layers.Flatten(),

    # 1st dense layer with increased dropout and LeakyReLU
    keras.layers.Dense(256, kernel_regularizer=l2(0.001)),
    LeakyReLU(),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.3),

    # 2nd dense layer
    keras.layers.Dense(128, kernel_regularizer=l2(0.001)),
    LeakyReLU(),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.3),

    # 3rd dense layer
    keras.layers.Dense(64, kernel_regularizer=l2(0.001)),
    LeakyReLU(),
    keras.layers.BatchNormalization(),
    keras.layers.Dropout(0.3),

    # output layer
    keras.layers.Dense(num_classes, activation='softmax')
])
```

```python
# Train the model
history = model_9.fit(X_train, y_train,
                      validation_data=(X_test, y_test),
                      batch_size=32,
                      epochs=100,
                      callbacks=[lr_scheduler, early_stopping, checkpoint])
```

```python
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization, Dropout, LeakyReLU
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping, ModelCheckpoint
from tensorflow.keras.regularizers import l2
import tensorflow as tf
```

```python
# Constructing the model
model_10 = tf.keras.Sequential([

    # Assuming MFCCs, start with convolutional layers
    tf.keras.layers.Reshape((X.shape[1], X.shape[2], 1), input_shape=(X.shape[1], X.shape[2])),
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4), # Increase dropout for convolutional layer

    # Flattening the output to feed into dense layers
    tf.keras.layers.Flatten(),

    # 1st dense layer with increased dropout and LeakyReLU
    tf.keras.layers.Dense(128, kernel_regularizer=l2(0.001)), # Reduced neurons
    LeakyReLU(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4), # Increased dropout

    # 2nd dense layer
    tf.keras.layers.Dense(64, kernel_regularizer=l2(0.001)),
    LeakyReLU(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),

    # output layer
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
```

# model 11

```python
model_11 = tf.keras.Sequential([

    # Reshape input for convolutional layers
    tf.keras.layers.Reshape((X.shape[1], X.shape[2], 1), input_shape=(X.shape[1], X.shape[2])),

    # 1st Convolutional layer
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D((2, 2)),
    tf.keras.layers.Dropout(0.4),

    # 2nd Convolutional layer
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.AveragePooling2D((2, 2)),
    tf.keras.layers.Dropout(0.4),

    # Global Average Pooling layer
    tf.keras.layers.GlobalAveragePooling2D(),

    # 1st Dense layer
    tf.keras.layers.Dense(128, kernel_regularizer=l2(0.001)),
    LeakyReLU(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),

    # 2nd Dense layer
    tf.keras.layers.Dense(64, kernel_regularizer=l2(0.001)),
    LeakyReLU(),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.4),

    # Output layer
    tf.keras.layers.Dense(num_classes, activation='softmax')
])
```

```python
from tensorflow.keras.models import load_model

# Assuming you have X_train, X_test, y_train, and y_test loaded in your environment
# If not, you should load them before evaluating the models.

def display_model_performance(model, model_name, X_train, y_train, X_test, y_test):
    train_loss, train_accuracy = model.evaluate(X_train, y_train, verbose=0)
    val_loss, val_accuracy = model.evaluate(X_test, y_test, verbose=0)

    print(f"Performance of {model_name}:")
    print(f"Training Accuracy: {train_accuracy:.4f}")
    print(f"Training Loss: {train_loss:.4f}")
    print(f"Validation Accuracy: {val_accuracy:.4f}")
    print(f"Validation Loss: {val_loss:.4f}")
    print("--------------------------------------\n")

models = {
    "Model_1": loaded_model_1,
    "Model_2": loaded_model_2,
    "Model_3": loaded_model_3,
    "Model_4": loaded_model_4,
    "Model_5": loaded_model_5,
    "Model_6": loaded_model_6,
    "Model_7": loaded_model_7,
    "Model_8": loaded_model_8,
    "Model_9": loaded_model_9,
    "Model_10": loaded_model_10,
    "Model_11": loaded_model_11,
}

for name, model in models.items():
    display_model_performance(model, name, X_train, y_train, X_test, y_test)
```

To determine the best model, we need to consider multiple factors:

1. **Validation Accuracy**: It tells us about the proportion of the predictions our model got right on unseen data. Higher is generally better.
2. **Validation Loss**: It indicates how well the model's predictions match the actual labels. Lower is better.
3. **Difference between Training and Validation Metrics**: A large gap between training and validation metrics can indicate overfitting.

Let's evaluate the models based on the provided metrics:

**Top Models based on Validation Accuracy**:
1. **Model_9** has the highest validation accuracy of 0.8989.
2. **Model_10** follows closely with a validation accuracy of 0.8962.
3. **Model_8** has the third-highest validation accuracy of 0.8932.

**Top Models based on Validation Loss**:
1. **Model_10** has the lowest validation loss of 0.6591.
2. **Model_11** has the second-lowest validation loss of 0.6022 (but a relatively lower accuracy compared to Model_10).
3. **Model_9** has the third-lowest validation loss of 0.7187.

**Considering Overfitting**:
When looking at the difference between training and validation metrics (to detect overfitting):
1. **Model_8** has a small gap between training and validation accuracy (0.9059 vs. 0.8932), suggesting that it is not severely overfitting.
2. **Model_9** and **Model_10** also have relatively small gaps, indicating they are not overfitting much either.

**Conclusion**:
- **Model_9** seems to be the best model, as it has the highest validation accuracy and a relatively low validation loss. The difference between its training and validation metrics isn't vast, indicating a balanced model.
- **Model_10** could be considered a close second given its performance metrics.
